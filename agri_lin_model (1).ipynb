```r
#load agri data for agriculture,forest,fishing and hunting

# independent variable --> 1. Price of feed 2. Agricultural price index 3. Price of fertilizer, 
# 4.Per capita meat consumption,
# 5. Trade-weighted index

# dependent variable --> domestic_demand

agri_df <- read.csv("Agribusiness_us_industry.csv")
```

```r
head(agri_df)
```

```r
summary(agri_df)
sapply(agri_df, sd)
cor(agri_df)
```

```r
plot(agri_df)
```

```r
#Creating a full model (a model with all variables included, without too much consideration).
model.full = lm(domestic_demand ~ ., data = agri_df)
```

```r
summary(model.full)
print("--------ANOVA-----------")
summary(aov(model.full))
```

```r

plot(model.full,1)
plot(model.full,2)
```

 An added-variable plot is the multivariable analogue of using a simple scatterplot
with a regression fit when there are no other covariates to show the relationship between
a single x variable and a y variable. An added-variable plot is a visually compelling
method for showing the nature of the partial correlation between x1 and y as estimated
in a multiple regression.


```r
avPlots(model.full)
# few parameters have either zero correlation ( price of fertilizer, price feed) 
#or its already being expressed by other variables 
```

```r
library(car)
vif(model.full)
```

```r
# we see very high vifs for year , price of feed index and agri_price_idx . 
# we remove 2 elements year and , price of feed idx
model1 = lm(domestic_demand ~ .   -Year - price_of_feed_Index  , data = agri_df)
summary(model1)
print("--------ANOVA-----------")
summary(aov(model1))
```

```r
#plot model1
plot(model1,1)
plot(model1,2)
```

```r
avPlots(model1)
```

```r
vif(model1)
```

#### price of fertilizer vif is pretty high we remove it and remodel it 


```r
# only one variable has vif < 5 
# model fails assumptions
#remove variable with highest vif value


model2 = lm(domestic_demand ~ .   -Year - price_of_feed_Index - price_of_fertilizer_Index  , data = agri_df)
#model2 = lm(domestic_demand ~ .   -Year  -agri_price_idx  , data = agri_df)
summary(model2)
print("--------ANOVA-----------")
summary(aov(model2))
```

```r

plot(model2,1)
plot(model2,2)
```

```r
vif(model2)
```

```r
avPlots(model2)
```

```r
anova(model2,model.full)
# p value is greater than 0.05 , we will fail to reject the null hypothesis.
#hence removed parameters the regression model do not significantly improve the fit of the model.
```
